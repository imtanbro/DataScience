```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv("flights_data.csv")
print(df.shape)
df.head()
```

```python
df.columns
```

# Bar Charts

```python
sb.countplot(data= df, x = "Source")
plt.ylabel('Number of Flights', fontsize = 12)
plt.xlabel('Source', fontsize = 12)
plt.show()
```

```python
base_color = sb.color_palette()[0]
sb.countplot(data=df, x = 'Source', color=base_color)
plt.xticks(rotation = 30)
plt.show()
```

```python
base_color = sb.color_palette()[3]
gen_order = df["Source"].value_counts().index
sb.countplot(data = df, x="Source", color=base_color, order=gen_order)
plt.show()
```

```python
base_color = sb.color_palette()[6]
sb.countplot(data=df, x = "Airline", color=base_color)
plt.show()
```

```python
base_color = sb.color_palette()[5]
sb.countplot(data=df, x = "Airline", color=base_color)
plt.xticks(rotation = 90)
plt.show()
```

```python
base_color = sb.color_palette()[9]
sb.countplot(data=df, y = "Airline", color=base_color)
plt.show()
```

# Counting Null/Missing data

```python
df.isna().sum()
```

```python
na_count = df.isna().sum()
base_color = sb.color_palette()[8]
sb.barplot(na_count.index.values, na_count, color=base_color)
plt.xticks(rotation = 90)
plt.ylabel("Number of missing values", fontsize = 12)
plt.show()
```

# Pie Chart

```python
sort = df['Destination'].value_counts()
plt.pie(sort, labels= sort.index, startangle=90, counterclock=False)
plt.axis('square')
plt.title('Flight Destination\'s')
plt.show()
```

```python
plt.pie(sort, labels= sort.index, startangle=90, counterclock=False, wedgeprops={'width':0.4})
plt.axis('square')
plt.title('Flight Destination\'s')
plt.show()
```

# Histograms

```python
plt.hist(data=df, x = "Duration(minutes)")
plt.show()
```

```python
plt.hist(data=df, x="Price", bins = 20)
plt.show()
```

```python
bins = np.arange(0,df['Price'].max()+1,1200)
plt.hist(data=df, x="Price", bins = bins)
plt.show()
```

```python
sb.distplot(df['Price'])
plt.show()
```

```python
sb.distplot(df['Price'],kde=False)
plt.show()
```

```python
bin_edges = np.arange(0,df['Price'].max()+1,1200)
sb.distplot(df['Price'],bins=bin_edges, kde=False, hist_kws={'alpha':1})
plt.show()
```
